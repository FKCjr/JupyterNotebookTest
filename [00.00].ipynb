So far we have only plotted prices against time, but what if we want to plot the density of the data and which values are like to occur. 

QQPlot : Quartile-Quantile Plot

```python
import scipy.stats
import pylab
```

```python
scipy.stats.probplot(df_comp.spx, plot = pylab)
pylab.show()
```

Theoretical quantiles : How many standard deviations away from the mean these values are. 

The straight orange line represents what the data should follow if the varialbles are normally distrubted however we can see that there are more values concenctrated on the 500 mark. Therefore, we cannot use the elegant statistics of normal distributions as the data is not normal to make successful forecasts. 

However this is what we expect from a time series
